```python
def base10_a_base12(n):
    if n == 0:
        return "0"

    digitos = "0123456789AB"  # Base 12 usa dígitos del 0 al 9 y A, B
    resultado = ""

    while n > 0:
        residuo = n % 12
        resultado = digitos[residuo] + resultado
        n //= 12

    return resultado

try:
    numero = int(input("Ingrese un número entero en base 10: "))
    if numero < 0:
        print("Por favor ingrese un número positivo.")
    else:
        convertido = base10_a_base12(numero)
        print(f"El número {numero} en base 12 es: {convertido}")
except ValueError:
    print("Entrada inválida. Ingrese un número entero.")



def base6_a_base10(num_base6):
    for digito in num_base6:
        if digito not in "012345":
            return None  # Número inválido en base 6

    num_base6 = num_base6[::-1]  # Invertir para empezar desde la posición 0
    resultado = 0

    for i in range(len(num_base6)):
        resultado += int(num_base6[i]) * (6 ** i)

    return resultado

numero = input("Ingrese un número en base 6: ")

resultado = base6_a_base10(numero)

if resultado is None:
    print("Error: el número ingresado no es válido en base 6.")
else:
    print(f"El número {numero} en base 10 es: {resultado}")
```
